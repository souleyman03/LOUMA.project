**Importer le fichier weekly reporting**

```python

# Charger la feuille nommée " "
df = pd.read_excel("LOUMAS REPORTING SEM14.xlsx", sheet_name="VTES VTO LOUMAS")
df.head()
```

```python
# Regrouper par direction, PVT, VTO
df_summary = df.groupby(['AGENCE_VENDEUR', 'ACCUEIL_VENDEUR', 'PRENOM_VENDEUR', 'NOM_VENDEUR', 'LOGIN_VENDEUR']).agg({
    'MSISDN': 'count'}).reset_index()
```

```python
df_summary
```

```python
# Trier les données pour regrouper visuellement
df_summary = df_summary.sort_values(['AGENCE_VENDEUR', 'ACCUEIL_VENDEUR'])

# Pour masquer les répétitions (laisser vide sauf première occurrence)
df_summary['AGENCE_VENDEUR'] = df_summary['AGENCE_VENDEUR'].mask(df_summary['AGENCE_VENDEUR'].duplicated())
df_summary['ACCUEIL_VENDEUR'] = df_summary['ACCUEIL_VENDEUR'].mask(df_summary['ACCUEIL_VENDEUR'].duplicated())
```

```python
df_summary
```

```python
# Renommer les colonne pour plus de clarté
df_summary = df_summary.rename(columns={'MSISDN': 'TOTAL_SIM_VENDUES'})
df_summary = df_summary.rename(columns={'ACCUEIL_VENDEUR': 'PVT'})
df_summary = df_summary.rename(columns={'LOGIN_VENDEUR': 'LOGIN'})
df_summary = df_summary.rename(columns={'AGENCE_VENDEUR': 'DRV'})
```

```python
df_summary
```

```python
# Regrouper par direction, PVT, VTO
df_summary2 = df.groupby(['AGENCE_VENDEUR', 'ACCUEIL_VENDEUR']).agg({
    'MSISDN': 'count'}).reset_index()
df_summary2 = df_summary2.rename(columns={'MSISDN': 'TOTAL_SIM_VENDUES'})
df_summary2 = df_summary2.rename(columns={'ACCUEIL_VENDEUR': 'PVT'})
df_summary2 = df_summary2.rename(columns={'AGENCE_VENDEUR': 'DRV'})
```

```python
df_summary2
```

```python

# Exporter les deux DataFrames dans un seul fichier Excel avec deux feuilles
with pd.ExcelWriter("WEEKLY_REPORTING.xlsx", engine="openpyxl") as writer:
    df_summary.to_excel(writer, sheet_name="RESUME VTE", index=False)
    df_summary2.to_excel(writer, sheet_name="VTE PVT", index=False)

print("✅ Fichier Excel généré avec 2 feuilles : RESUME VTE & VTE PVT")
```
